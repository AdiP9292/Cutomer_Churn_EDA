```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

converted the Senior citizen from 0 and 1 to yes/no

```python
def conv(value):
  if value == 1:
    return 'Yes'
  else:
    return 'No'
```

```python
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
ax = sns.countplot(x='Churn',data=df)

ax.bar_label(ax.containers[0])
plt.title("Count of the Churn Customers ")
plt.show()
```

```python
plt.figure(figsize=(3,4))
gb = df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"],labels = gb.index,autopct = "%.2f%%")
plt.title("Churn")
plt.show()
```

```python
plt.figure(figsize=(3,4))
bx = sns.countplot(x="SeniorCitizen",data =df,hue = "Churn")
plt.title("Senior Citizen Churn")
plt.show()
```

```python
plt.figure(figsize=(8, 6))  # Adjust figure size as needed

# Calculate churn percentages
churn_counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)
churn_percentages = churn_counts.div(churn_counts.sum(axis=1), axis=0) * 100

# Plotting the stacked bar chart
ax = churn_percentages.plot(kind='bar', stacked=True)

# Adding labels and title
plt.title("Senior Citizen Churn (%)")
plt.xlabel("Senior Citizen (0: No, 1: Yes)") # Added x-axis label for clarity
plt.ylabel("Churn Percentage") # Added y-axis label

# Adding percentage labels on the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height:.1f}%', (x + width/2, y + height/2), ha='center', va='center', fontsize=9, color='black')  # Adjust fontsize as needed

# Improve Legend
plt.legend(title='Churn', bbox_to_anchor=(1.05, 1), loc='upper left') # Place legend outside the plot

plt.xticks(rotation=0) # Keep x-axis labels horizontal if needed

plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()

```

```python
columns_to_plot = ['PhoneService', 'MultipleLines', 'InternetService',
                   'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                   'TechSupport', 'StreamingTV', 'StreamingMovies']

num_cols = 3  # Number of columns in the subplot grid
num_rows = (len(columns_to_plot) + num_cols - 1) // num_cols # Calculate rows dynamically

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5*num_rows)) # Adjust figsize as needed
fig.suptitle("Count Plots of Categorical Features", fontsize=16) # Overall title

# Flatten the axes array for easier iteration
axes = axes.flatten()

for i, column in enumerate(columns_to_plot):
    sns.countplot(x=column, data=df, ax=axes[i],hue ="Churn")
    axes[i].set_title(column)
    axes[i].tick_params(axis='x') # Rotate x-axis labels if needed

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    axes[j].set_axis_off()

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent overlapping
plt.show()
```
